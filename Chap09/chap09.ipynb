#### 코드 9.26

```python
import pandas as pd
data = { 'Test1' : pd.Series([70, 55, 89],
        index=['Park', 'Kim', 'Han']),
        'Test2' : pd.Series([56, 82, 77, 65],
        index=['Park', 'Kim', 'Han', 'Lee'])}
df1 = pd.DataFrame(data)
print (df1)
df1['Project'] = pd.Series([90, 83, 67, 87],
                index=['Han','Kim','Lee', 'Park'])
print ("="*20)
df1['Average'] = round((df1['Test1']+df1['Test2']+
                        df1['Project'])/3, 2)
print (df1)
```

#### 코드 9.27

```python
df2 = df1
del df2['Test2']
print (df2)
```

#### 코드 9.28

```python
df2 = df1
df2.pop("Project")
print (df2)
```

#### 코드 9.29

```python
df2= df1.copy() # copy df1 into df2 using copy() method
print (df2)
#delete columns using del and pop methods
del df2['Test2']
df2.pop('Project')
print("=== 열 삭제된 결과>>> \n", df2)
print("=== 원래의 DataFrame>>> \n", df1)
```

#### 코드 9.42

```python
def cleaning(x) :
  if str(x).find("?") != -1 :
    return str(x).split("?")[1].replace("?","") 
  return str(x)

summer.iloc[:, 0].apply(cleaning).apply(lambda x:x.replace(" ",""))
summer.head()
```
