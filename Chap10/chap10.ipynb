#### 코드 10.32

```python
import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0, 5, 10)

fig, ax1 = plt.subplots()
ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel("area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")

ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel("volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")

plt.show()
```

#### 코드 10.33

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, 5, 10)
y = x ** 2
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main[left,bottom,width,height]
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # inner axes
# main figure
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('outter_plot title')
# insert
axes2.plot(y, x, 'g')
axes2.set_xlabel('y')
axes2.set_ylabel('x')
axes2.set_title('inner_plot title');
plt.show()
```

#### 코드 10.35

```python
import numpy as np
import matplotlib.pyplot as plt

N = 5
menMeans = (40, 35, 30, 35, 37)
womenMeans = (42, 42, 34, 30, 35)

ind = np.arange(N)    # the x locations for the groups
width = 0.35          # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, menMeans, width, color='red')
p2 = plt.bar(ind, womenMeans, width, bottom=menMeans)

plt.ylabel('Scores')
plt.title('Scores by group and gender')
plt.xticks(ind, ('G1', 'G2', 'G3', 'G4', 'G5'))
plt.yticks(np.arange(0, 110, 10))
plt.legend((p1[0], p2[0]), ('Men', 'Women'))
plt.show()
```

#### 코드 10.44

```python
import matplotlib.pyplot as plt
import csv
fp = open('temp10.csv')
data = csv.reader(fp)
next(data)
Jan_July = []
for each in data :
    month = each[0].split('-')[1]
    if each [2] != '':
        if month == '01' or month == '07':
            Jan_July.append(float(each[2]))
plt.hist(Jan_July, bins=100, color='b')
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv
fp = open('temp10.csv')
data = csv.reader(fp)
next(data)
Aug = []
for each in data :
    month = each[0].split('-')[1]
    if each [-1] != '':
        if month == '08':
            Aug.append(float(each[-1]))
plt.boxplot(Aug)
plt.show()
```
