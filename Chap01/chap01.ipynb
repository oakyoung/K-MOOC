### Chapter 1. 데이터 분석

```python
import numpy as np
import csv
f =open('age.csv')
data = csv.reader(f)
next(data)
name = input('인구 구조가 알고 싶은 지역의 이름을 입력해주세요 : ')
for row in data :
    if name in row[0] :
        home = np.array(row[3:], dtype = int) 

import matplotlib.pyplot as plt
plt.style.use('ggplot')
plt.figure(figsize = (10,5), dpi=300)    
plt.rc('font', family ='Malgun Gothic')
plt.title(name +' 지역의 인구 구조')

plt.plot(home) 
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rc('font', family = 'Malgun Gothic')   #run command (한글 폰트 세팅)
df = pd.read_csv('age.csv', encoding = 'cp949', index_col=0)
df
```

```python
df = df.div(df['총인구수'], axis=0)
df
```

```python
del df['총인구수'], df['연령구간인구수']
```

```python
df
```

```python
location = input("지역을 선택해주세요 : ")
```

```python
select = df.index.str.contains(location)
df2 = df[select]
df2
```

```python
fig = plt.figure(figsize=(8,8), facecolor='white')
```

```python
df.loc[np.power(df.sub(df2.iloc[0], axis=1), 2).sum(axis=1).sort_values().index[:5]].T.plot()
plt.show()
```
