```python
with open('article_text.txt', 'r', encoding='CP949') as f:
    article_text = f.read()
```

#### 코드 13.23

```python
#Okt를 이용하여 책 소개 텍스트들의 형태소 분석
from konlpy.tag import Okt
import nltk
from collections import Counter

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')

okt = Okt()

k_words = okt.nouns(article_text) # 명사만 추출
c = Counter(k_words) # 위에서 얻은 words를 처리하여 단어별 빈도수 형태의 딕셔너리 데이터를 구함
tags = c.most_common(100) 
for each in tags:
    print(each)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import matplotlib.patches as patches

font = '/usr/share/fonts/truetype/nanum/NanumGothicEco.ttf'
wc = WordCloud(font_path=font, background_color='white',
               width=400, height=400, scale=2.0, max_font_size=250)
cloud = wc.generate_from_frequencies(dict(tags))

plt.figure(figsize=(8,8))
plt.axis('off')
plt.imshow(cloud)
```
