### 확률분석 도전
---
**문제 상황**

어떤 제빵회사에서 파는 빵에는 귀여운 스티커가 하나씩 들어 있다.

총 151가지 종류의 스티커가 있으며, 각 종류의 스티커가 나올 확률은 1/151로 모두 같다.

빵을 사서 뜯기 전에는 어떤 종류의 스티커가 들어있는지 알 수 없다.

스티커를 종류별로 하나이상씩 모두 모으려고 할 때, 총 몇 개의 빵을 먹어야 할까?

(1) 시뮬레이션

#### 코드 11.6

```python
import random
import numpy as np
```

```python
N = 151
M = 100000
res = []

for i in range(M):
    have = [0] * N #0~150번 스티커를 갖고 있는 상태인지 저장
    new = 0 #지금까지 모은 스티커 종류 수
    num = 0 #먹은 빵 개수
    while True:
        num += 1
        x = np.random.randint(N)
        if have[x] == 0: #새로운 스티커가 나왔을 때
            have[x] = 1
            new += 1
            if new == N: #N개 다 모았을 때
                res.append(num)
                if i% 1000 == 0: #1000번에 한 번씩 로그 찍기
                  print(f"{i+1}회 실행 결과:{num}개 빵으로 모든 스티커 획득!")
                break
```

#### 코드 11.7

```python
#10만번 반복 시행했을 때, 평균값
print(np.mean(res)) 
```

(2) 확률 계산

#### 코드 11.8

```python
#첫번째 스티커를 모으려 할 때, 총 먹어야 하는 빵 개수
p = N/N #첫번째 새로운 스티커가 나올 확률
print(1/p) #기하분포에서 시행횟수의 기대값은 확률의 역수
```

#### 코드 11.9

```python
#위 상태에서 두번째 스티커를 모으려 할 때, 총 먹어야 하는 빵 개수

p = (N-1)/N #두번째 새로운 스티커가 나올 확률
print(1/p) #먹어야 하는 빵 개수의 기대값
```

#### 코드 11.10

```python
#... 151번째 스티커를 모으려 할 때, 총 먹어야 하는 빵 개수

p = 1/N #마지막 스티커가 새로 나올 확률
print(1/p) #먹어야 하는 빵 개수의 기대값
```

#### 코드 11.11

```python
#전체 먹어야 하는 빵 개수의 기대값
total = 0
for i in range(N):
    p = (N-i)/N
    total += 1/p
print(total)
```

#### 코드 11.14

```python
# 이항분포의 확률질량함수(pmf) 시각화
n = 100
p = 10/151

# ppf: Percent point function, 분위수 구하기 => 1% ~ 99%의 k값을 x축의 값으로
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p)) 

# pmf: Probability mass function
plt.plot(x, binom.pmf(x, n, p), 'ko') 
plt.vlines(x, 0, binom.pmf(x, n, p), colors = 'brown')
plt.xlabel('k')
plt.ylabel('P(k)')
plt.title('Binomial distribution pmf')
plt.show()
```

#### 코드 11.15

```python
#위 이항분포의 누적분포함수(cdf) 시각화
n = 100
p = 10/151

# ppf: Percent point function, 분위수 구하기 => 1% ~ 99%의 k값을 x축의 값으로
x = np.arange(binom.ppf(0.01, n, p), binom.ppf(0.99, n, p)) 

#cdf: Cumulative distribution function
plt.plot(x, binom.cdf(x, n, p), 'ko') 
plt.vlines(x, 0, binom.cdf(x, n, p), colors = 'brown')
plt.xlabel('k')
plt.ylabel('P(X<=k)')
plt.title('Binomial distribution cdf')
plt.show()
```
