```python
sel_data = my_sheet['A']
my_sheet['B1'] = '주인공'
for data in sel_data:
    if data.value == '오징어 게임':
        sel_col, sel_row = data.coordinate  # 오징어 게임이 저장된 셀 주소 반환
        my_sheet['B'+sel_row] = '이정재'    # 해당 행에 주인공 정보 입력
        
        check_data = my_sheet[int(sel_row)]  # 저장된 내용 확인
        for each in check_data:
            print (each.value)
```

코드 5.30

```python
#openpyxl 라이브러리를 불러오고, 다뤄볼 엑셀 데이터를 불러오기
#데이터 출처: https://data.seoul.go.kr/dataList/OA-15069/F/1/datasetView.do
import openpyxl
workbook = openpyxl.load_workbook('서울특별시 가로쓰레기통 설치정보(2019.9).xlsx', data_only=True)
wb1 = workbook['상세현황(중앙차로제외)']

#출력하여 제대로 불러왔는지 확인하기
for row in wb1.rows:
  for cell in row:
    print(cell.value, end = ' ')
  print()
```

코드 5.31

```python
location = wb1['C']
sel_data = []
for data in location:
    if data.value == '삼청로':
        sel_row = data.coordinate[1:]    # 행 번호 선택
        pick = 'D'+ sel_row
        sel_data.append ( wb1[pick].value )   # 설치 위치 확인
for each in sel_data:
    print ( each )
```

코드 5.32

```python
from openpyxl import Workbook  
  
wb = Workbook()  
sheet = wb.active  
  
rows = (  
    ('A', 496, 48, 454),  
    ('B', 730, 42, 116),  
    ('C', 905, 17,257),  
    ('D', 132, 839, 453),  
    ('F', 81, 240, 444),  
    ('E', 151, 716, 482),
    ('G', 100, 200, 300)
)  
  
for row in rows:  
    sheet.append(row)  
  
wb.save('new_data.xlsx')  
```

코드 5.33

```python
wb_sheet = wb.active

# 특정 Column 숨기기
wb_sheet.column_dimensions.group('B','C', hidden=True)

# 특정 Row 숨기기
wb_sheet.row_dimensions.group(2, 4, hidden=True)

wb.save("updated_data.xlsx") #실행결과 확인 - 접혀서 숨겨진 것을 확인
```

코드 5.34

```python
workbook = openpyxl.load_workbook('new_data.xlsx')

wb_sheet = workbook.active

wb_sheet['A8'] = 'Total'

sum = 0 
data = wb_sheet['B']
for each in data:
    sum += int(each.value)

cell = wb_sheet.cell(row=8, column=2)  
cell.value = sum  

for row in wb_sheet.rows:
  for cell in row:
    print(cell.value, end = ' ')
  print()
```
