코드 4.1 <br>
실행시키는 시점의 뉴스이므로 결과물이 다릅니다. 

```python
# 웹페이지 요청을 위한 패키지
import requests

# 네이버 뉴스 페이지 url
url = 'https://news.naver.com/'

# 네이버 뉴스 페이지 요청, 응답 저장
response = requests.get(url)
print(response.text)  # 내용 확인
```

코드 4.2 

```python
# 웹페이지 파싱을 위한 패키지
from bs4 import BeautifulSoup

# response로 부터 html 가져오기
html = response.text

# BeautifulSoup 으로 html 파싱
soup = BeautifulSoup(html, 'html.parser')

# html에서 원하는 태그 찾기
hdline_titles = soup.find_all('div', {"class":"cjs_t"})

# div에 cjs_t의 class에 속한 텍스트만 추출
for title in hdline_titles:
    print(title.get_text(strip=True))   # 결과 출력
```

코드 4.4

```python
word_list = text.split()

count_list = [ ]
for word in word_list :
    count = word_list.count(word)   # 한 단어씩 빈도수 계산
    count_list.append ( [word, count] )

# 빈도수 높은 순으로 정렬
sorted_list = sorted(count_list, key = lambda x:x[1], reverse=True)

# 중복 자료 제거 후 출력
word_dict = {}
for word, frequency in sorted_list:
    if word not in word_dict:
        word_dict[word] = frequency
        print(word, frequency)
        
```

코드 4.6 <br>
ppt 파일은 다른 파일을 사용하시기 바랍니다. <br>
사용하신 파일명으로 수정하신 후 실행하세요.

```python
from pptx import Presentation

contents = Presentation("데이터이해와활용.pptx")
text_collection = []
for slide in contents.slides:
    for shape in slide.shapes:
        if not shape.has_text_frame:
            continue
        for paragraph in shape.text_frame.paragraphs:
            text_collection.append(paragraph.text)
print(text_collection)
```

코드 4.7

```python
# text_data.txt 파일에 저장하기
outfp = open ("text_data.txt", 'w', encoding = 'utf-8')
outfp.writelines (text_collection)
outfp.close()

# 저장된 자료 확인
infp = open ("text_data.txt", 'r', encoding = 'utf-8')
data = infp.read()
print(data)
```

코드 4.12 

```python
#BeautifulSoup 불러오기
import requests 
from bs4 import BeautifulSoup
```

```python
#크롤링할 웹사이트 설정
url = 'https://www.korea.kr/news/top50List.do'
resp = requests.get(url)

soup = BeautifulSoup(resp.text)

top_news = []

#크롤링할 태그 조건 설정
strong = soup.select("span.text > strong") # text라는 클래스명을 가진 span 태그의 자식 중 strong 태그만 선택
for n in strong:
  top_news.append(n.get_text().strip())

#크롤링 결과 확인
print(top_news)
```

코드 4.14

```python
#텍스트 파일 불러오기
file_name = "먼 후일.txt"
fp = open(file_name, "r", encoding = "utf-8") #읽기용

#파일 읽기
data = fp.readlines() #리스트로 저장
fp.close() 

print(data)
```

코드 4.15

```python
#한 줄씩 출력
for line in data: 
    print(line, end='')
```

코드 4.16

```python
#복사본 파일 생성하고 열기
fp2 = open(file_name.split(".")[0]+"_작업본.txt","w") #쓰기용

cnt=0
where=[]

for i,line in enumerate(data): 
    if "잊었노라" in line : #특정 단어 검색
        cnt+=1
        where.append(i)
        fp2.write(str(i) + ' : ' + line) #한 줄씩 쓰기
    
fp2.close()

print(f"'잊었노라'는 {where}줄에 총 {cnt}번 존재합니다.")
```
